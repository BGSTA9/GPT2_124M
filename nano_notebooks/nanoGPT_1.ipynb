    Now let's implement some of the skeleton of the "nn.Module" in the "train_gpt2.py" file, and we will do our best to match up the schema for the HuggingFace, we saw earlier.

    The main container which had all the modules was called "transfomer." In our code we will be reflecting that with the nn.ModuleDict. And this is basically a module that allows you to index into submodules using a keys, just like a dictionary strings.

        The following is the skeleton that reflects the image above.

    within our container we have the following submodules:
    - wte: Weights of the Token Embeddings, "nn.Embedding".
    - wpe: Weight of the Position Embedding, we also have the "nn.Embedding".
        -> "nn.Embedding" is a fancy wrapper around a single array of numbers, block of numbers, just like the gray pattern we saw through 'plt'. So this allows us to access its elements, by indexing into the rows.
    - h: hidden layers. The 12 layers of the transformer, and instead of using the "nn.ModuleDict" we use the "nn.ModuleList", so we be able to index using the integer, exactly like the "h" in the HuggingFace.
    - ln_f: the additional final "LayerNorm"
    - lm_head: the final "Linear" layer classifier, that project from 'n_embd' all the way to the 'vocab_size', with using no bias [bias=False] for the final projection.
